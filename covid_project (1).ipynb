<h1>Now date wise growth anlysis</h1>

```python

date_wise = all_data.groupby(all_data.Date).agg({"Confirmed":'sum',"Recovered":'sum',"Deaths":'sum',"Active":'max'})
for i,x in enumerate(date_wise.index):
  if i ==0 :
    pass
  else:
    if date_wise.Confirmed[i] < date_wise.Confirmed[i-1]:
      date_wise.Confirmed[i] =  (date_wise.Confirmed[i-1]+ date_wise.Confirmed[i+1])/2
      # print("changing {} to {}".format(date_wise.Confirmed[i],date_wise.Confirmed[i-1]))
    if date_wise.Recovered[i] < date_wise.Recovered[i-1]:
      date_wise.Recovered[i] = (date_wise.Recovered[i-1]+date_wise.Recovered[i+1])/2
    if date_wise.Deaths[i] < date_wise.Deaths[i-1]:
      date_wise.Deaths[i] = (date_wise.Deaths[i-1]+date_wise.Deaths[i+2])/2

date_wise["Recovery Rate"] = date_wise["Recovered"] /date_wise["Confirmed"]*100
date_wise["Death Rate"] = date_wise["Deaths"] /date_wise["Confirmed"]*100
newcases = []
newdeath = []
newrecovered = []
for i,x in enumerate(date_wise.Confirmed):
  if i == 0:
    newcases.append(date_wise.Confirmed[i])
    newdeath.append(date_wise.Deaths[i])
    newrecovered.append(date_wise.Recovered[i])
  else:
    newcases.append(date_wise.Confirmed[i] - date_wise.Confirmed[i-1])
    newdeath.append(date_wise.Deaths[i] - date_wise.Deaths[i-1])
    newrecovered.append(date_wise.Recovered[i] - date_wise.Recovered[i-1])

date_wise["New Cases"] = newcases
date_wise["New Recovered"] = newrecovered
date_wise["New Deaths"] = newdeath

```

<h1>Month wise analysis of india</h1>

```python
#Month Wise analysis of india 

india_data["month"] = india_data.index.month_name()
# india_data["month"] = india_data["month"].astype(str).apply(lambda x: dt.datetime.strptime(x,'%b')) 
india_data["dailyconfirmed"] = india_data["dailyconfirmed"].astype(int)
india_data["totalconfirmed"] = india_data["totalconfirmed"].astype(int)
india_data["totalrecovered"] = india_data["totalrecovered"].astype(int)
india_data["totaldeceased"] = india_data["totaldeceased"].astype(int)
month_wise = india_data.groupby("month").agg({"totalconfirmed":"max","totalrecovered":"max","totaldeceased":"max","dailyconfirmed":"sum"})
month_wise["Recovery Rate"] = month_wise["totalrecovered"].astype(int) /month_wise["totalconfirmed"].astype(int) *100
month_wise["Death Rate"] = month_wise["totaldeceased"].astype(int)  /month_wise["totalconfirmed"].astype(int) *100
month_wise["Growth rate"] = month_wise["dailyconfirmed"].astype(int) /month_wise["totalconfirmed"].astype(int) *100
month_wise  = month_wise.sort_values("totalconfirmed",ascending=True)
```

```python
month_wise.head(7)
```

```python
fig = make_subplots(rows=3, cols=1,
                   subplot_titles=("Total Confirmed ", "Total Recovered","Total Deaths"))
fig.add_trace(
    go.Scatter(x=month_wise.index, y=month_wise["totalconfirmed"],name="Total Confirmed"),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=month_wise.index, y=month_wise["totalrecovered"],name="Total Recovered"),
    row=2, col=1
)
fig.add_trace(
    go.Scatter(x=month_wise.index, y=month_wise["totaldeceased"],name="Total Deaths"),
    row=3, col=1
)
fig.update_layout(height=1300,legend=dict(x=-0.1,y=1,traceorder="normal"))
fig.show()
```

```python
fig = make_subplots(rows=3, cols=1,
                   subplot_titles=("Growth rate ", "Death Rate","Recovery rate"))
fig.add_trace(
    go.Scatter(x=month_wise.index, y=month_wise["Growth rate"],name="Growth Rate"),
    row=1, col=1
)
fig.add_trace(
    go.Scatter(x=month_wise.index, y=month_wise["Death Rate"],name="Death Rate"),
    row=2, col=1
)
fig.add_trace(
    go.Scatter(x=month_wise.index, y=month_wise["Recovery Rate"],name="Recovery Rate"),
    row=3, col=1
)

fig.update_layout(height=1300,legend=dict(x=-0.1,y=1.1,traceorder="normal"))
```
